```python
import json
```

```python
with open('levelData.json', 'r') as json_file:
    data = json.load(json_file)
```

```python
for room in data.values():
    print("ROOM: " + str(room["RoomIndex"]))
    for wall in room["Walls"]:
        print(wall["Max"])
        print(wall["Min"])
        for area in wall["Path"]:
            print(json.dumps(area, indent = 4))
```
